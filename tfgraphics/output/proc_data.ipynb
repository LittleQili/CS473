```python
import numpy as np 
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import math
import random
```

## 原始代码

```python
loss_epoch_init = []
loss_val_init = []
eval_epoch_init = []
eval_val_init = []
with open("naiveoutput.txt", "r") as fr:  # 打开文件
    readdata = fr.read()  # 读取文件
    # print(readdata)
# type(readdata)
linedata = readdata.split('\n')
for line in linedata:
    # print("one line: ",line)
    if line.find('INFO:tensorflow:loss = ') != -1:
        tmplossline = line.replace('INFO:tensorflow:loss = ','')
        tmplossline = tmplossline.replace(' step = ','')
        tmplossline = tmplossline.replace(' (',',')
        tmplossline = tmplossline.replace(' sec)','')
        numlist = tmplossline.split(',')
        loss_val_init.append(float(numlist[0]))
        loss_epoch_init.append(float(numlist[1]))
        # INFO:tensorflow:Saving dict for global step 
    if line.find('INFO:tensorflow:Saving dict for global step ') != -1:
        tmplossline = line.replace('INFO:tensorflow:Saving dict for global step ','')
        tmplossline = tmplossline.replace(': accuracy = ',',')
        numlist = tmplossline.split(',')
        eval_epoch_init.append(float(numlist[0]))
        eval_val_init.append(float(numlist[1]))
# print(eval_epoch)
# eval_val
```

```python
sns.set_style("darkgrid")
fig, ax = plt.subplots(1,2,figsize=(20,5))

ax[0].plot(loss_epoch_init,loss_val_init)
ax[0].set_title("Initial Loss",fontsize=15)
ax[1].plot(eval_epoch_init,eval_val_init)
ax[1].set_title("Initial Accuracy",fontsize=15)
```
